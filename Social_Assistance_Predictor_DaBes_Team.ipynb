# Preprocessing Data

# By Year

```python
garis_kemiskinan = garis_kemiskinan_perkapita[garis_kemiskinan_perkapita['tahun'] == 2019]
# garis_kemiskinan
```

# DataFrame to List

```python
kabupaten_kota = list(x for x in bansos_2019['label'])

jumlah_keluarga_rencana_2018 = list(x for x in bansos_2018['kementerian_sosial_rencana_jumlah_keluarga_penerima_manfaat_kpm_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_keluarga_realisasi_2018 = list(x for x in bansos_2018['kementerian_sosial_realisasi_jumlah_keluarga_penerima_manfaat_kpm_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_anggaran_rencana_2018 = list(x for x in bansos_2018['kementerian_sosial_rencana_anggaran_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_anggaran_realisasi_2018 = list(x for x in bansos_2018['kementerian_sosial_realisasi_anggaran_bantuan_sosial_pangan_bansos_pangan_'])

jumlah_keluarga_rencana_2019 = list(x for x in bansos_2019['kementerian_sosial_rencana_jumlah_keluarga_penerima_manfaat_kpm_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_keluarga_realisasi_2019 = list(x for x in bansos_2019['kementerian_sosial_realisasi_jumlah_keluarga_penerima_manfaat_kpm_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_anggaran_rencana_2019 = list(x for x in bansos_2019['kementerian_sosial_rencana_anggaran_bantuan_sosial_pangan_bansos_pangan_'])
jumlah_anggaran_realisasi_2019 = list(x for x in bansos_2019['kementerian_sosial_realisasi_anggaran_bantuan_sosial_pangan_bansos_pangan_'])

kode = list(x for x in rumah['kode_kabupaten_kota'])
kode.append(32)

rumah_tidak_layak_huni = list(x for x in rumah['jumlah_pemilik_rutilahu'])
rumah_tidak_layak_huni.append(sum(rumah_tidak_layak_huni))

kemiskinan = list(x for x in garis_kemiskinan['garis_kemiskinan_perkapita'])
kemiskinan.append(round(sum(kemiskinan)/len(kemiskinan)))

tahun = list(2019 for x in range(28))
```

# Convert Data to Numeric

```python
def to_integer(data):
    a = []
    for i in data:
        b = []
        for j in i:
            if j.isnumeric():
                b.append(j)
            elif j == ",":
                break
        a.append(int("".join(b)))
    return a
```

```python
int_keluarga_rencana_2018 = to_integer(jumlah_keluarga_rencana_2018)
int_keluarga_realisasi_2018 = to_integer(jumlah_keluarga_realisasi_2018)
int_anggaran_rencana_2018 = to_integer(jumlah_anggaran_rencana_2018)
int_anggaran_realisasi_2018 = to_integer(jumlah_anggaran_realisasi_2018)

int_keluarga_rencana_2019 = to_integer(jumlah_keluarga_rencana_2019)
int_keluarga_realisasi_2019 = to_integer(jumlah_keluarga_realisasi_2019)
int_anggaran_rencana_2019 = to_integer(jumlah_anggaran_rencana_2019)
int_anggaran_realisasi_2019 = to_integer(jumlah_anggaran_realisasi_2019)
```

# New DataFrame for Analysis

```python
df_bansos = pd.DataFrame(kabupaten_kota , columns= ['Kabupaten/Kota'])

df_bansos['Kode'] = kode

df_bansos['Garis Kemiskinan Perkapita (Rp)'] = kemiskinan

df_bansos['Rumah Tidak Layak Huni (KK)'] = rumah_tidak_layak_huni

df_bansos['Indeks Kedalaman Kemiskinan (P1)'] = indeks_kedalaman['2019']

df_bansos['Indeks Keparahan Kemiskinan (P2)'] = indeks_keparahan['2019']

# df_bansos['Jumlah Keluarga 2018 (Rencana)'] = int_keluarga_rencana_2018
# df_bansos['Jumlah Keluarga 2018 (Realisasi)'] = int_keluarga_realisasi_2018
# df_bansos['Jumlah Anggaran 2018 (Rencana)'] = int_anggaran_rencana_2018
# df_bansos['Jumlah Anggaran 2018 (Realisasi)'] = int_anggaran_realisasi_2018

df_bansos['Rencana Jumlah Keluarga (KK)'] = int_keluarga_rencana_2019
df_bansos['Realisasi Jumlah Keluarga (KK)'] = int_keluarga_realisasi_2019
df_bansos['Rencana Anggaran (Rp)'] = int_anggaran_rencana_2019
df_bansos['Realisasi Anggaran (Rp)'] = int_anggaran_realisasi_2019

df_bansos['Tahun'] = tahun
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split

df_feature = df_model[['Garis Kemiskinan Perkapita (Rp)', 
                      'Rumah Tidak Layak Huni (KK)', 
                      'Rencana Jumlah Keluarga (KK)']]

df_target = df_model[['Realisasi Anggaran (Rp)']]

X_train, X_test, y_train, y_test = train_test_split(df_feature, df_target, test_size=0.15, random_state=0)

print(f'X_train : {X_train.size} (85%) | X_test : {X_test.size} (15%)')
print(f'y_train : {y_train.size} (85%) | y_test : {y_test.size}  (15%)')
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler

preprocessor = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = preprocessor.fit_transform(X_train)
y_train_scaled = preprocessor.fit_transform(y_train)
```

# Train Model

```python
# Before Scaling
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print(f"Konstanta (a) : {model.intercept_}")
print(f"Koefisien (b) : {model.coef_}")
```

```python
# After Scaling
from sklearn.linear_model import LinearRegression

model2 = LinearRegression()
model2.fit(X_train_scaled, y_train_scaled)

print(model2.coef_)
print(model2.intercept_)
```
